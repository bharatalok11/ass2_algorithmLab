```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ObservationDPcpp.csv')
df
```

```python
print(df.columns)
```

```python
vertices = df['Vertices']
avg_time_taken_brute = df['Avg.Time Taken']
avg_time_taken_dp = df['Avg.Time Taken DP']
```

```python
plt.figure(figsize=(10, 6))
plt.plot(vertices, avg_time_taken_brute, label='Avg. Time Taken (Brute Force)', marker='o')
plt.plot(vertices, avg_time_taken_dp, label='Avg. Time Taken (DP)', marker='x')
plt.xlabel('Number of Vertices')
plt.ylabel('Average Time Taken (ms)')
plt.title('Comparison of Average Time Taken for DP approach')
plt.legend()
plt.show()
```
