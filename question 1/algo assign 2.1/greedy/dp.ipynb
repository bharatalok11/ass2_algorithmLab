```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ObservationDP1.csv')
df
```

```python
print(df.columns)
```

```python
vertices = df['Vertices']
avg_time_taken = df['Avg.Time Taken']
```

```python
plt.figure(figsize=(10, 6))
plt.plot(vertices, avg_time_taken, label='Avg. Time Taken (DP)', marker='o')
plt.xlabel('Number of Vertices')
plt.ylabel('Average Time Taken (ms)')
plt.title('Comparison of Average Time Taken for Dynamic Programming approach')
plt.legend()
plt.show()
```
