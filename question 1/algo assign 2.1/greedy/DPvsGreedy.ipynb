```python
vertices = df['Vertices']
avg_time_taken_dp = df[' Avg.Time Taken DP']
avg_time_taken_greedy = df[' Avg.Time Taken Greedy']
```

```python
plt.figure(figsize=(10, 6))
plt.plot(vertices, avg_time_taken_dp, label='Avg. Time Taken (DP)', marker='o')
plt.plot(vertices, avg_time_taken_greedy, label='Avg. Time Taken (Greedy)', marker='x')
plt.xlabel('Number of Vertices')
plt.ylabel('Average Time Taken (ms)')
plt.title('Comparison of Average Time Taken for Triangulation')
plt.legend()
plt.show()
```
